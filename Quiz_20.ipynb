### **Quiz 20**
For each pair of characters of an input String, swap the two characters. 

```
# Create a class solution to solve the question
class Solution():

  def __init__(self, input):
    # initialize the input as a string
    self.input: str = input
# Create a function to swap every two characters in an input string
  def swap_pair(self):
    input = self.input
    # create a string to store the final output
    out = ""
    # situation 1: if the input string only has one or no element, there's no need to swap every two characters
    if len(input) < 2:
      return input
    # situation 2: if the input string has at least two strings, then it can swap
    else:
      # loop in the input string and takes the elements two by two
      for i in range(0,len(input),2):
        # swap the order of the two consecutive elements
        out += input[i+1] + input[i]
      return out

print(Solution("monkey").swap_pair())
print(Solution("A").swap_pair())
```
