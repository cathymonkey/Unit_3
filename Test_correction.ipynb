````
## TEST for question f
neuron = [1,2,3,4] 
cur_output = [0.5,0.2,0.2,-0.1]
num_inputs = [1,2,2,1]
strength = [-0.1,0.5,-0.4,0.2]
source = [4,1,3,1,2,1]

def get_index(id):
  index = -1
  for i in neuron:
    index += 1
    if i == id:
      return index
print("Question d output:")
get_index(2)
  
def simulate(neuron,cur_output,num_inputs,strength,source):
  if check_inputs():
    start_idx = 0
    for n in range(start_idx,len(neuron)-1,1):
      end_idx = start_idx + num_inputs[n]
      out = 0.0
      for k in range(start_idx,end_idx-1,1):
        temp = get_index(source[k])
        out += strength[temp]*cur_output[temp]
      cur_output[n] = out
      start_idx = end_idx
      print("Output of Neuron id",neuron[n],"is",out)
    return cur_output

  else:
      return "Error in the initialization"

print("Question f output:")
simulate(neuron,cur_output,num_inputs,strength,source)

````
