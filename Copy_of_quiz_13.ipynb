Given an array of food items and price, calculate the total price based on the table of tax categories below.

| Item        | Tax rate |
|-------------|----------|
| Food        | 10%      |
| Electronics | 15%      |
| Liqueur     | 20%      |

Screen Shot 2021-03-03 at 10.11.24.png

```
# goods list by category
food = ["bread","egg"]
electronics = ["ipad","Kindle"]
liqueur = ["beer"]
# Create a function total to calculate the sum of the price and tax of the order
def total(item,price):
# create a variable price_sum for adding the price and tax of the goods later on
    price_sum = 0

# loop in the item list 
    for i in range(len(item)):
      # check which category the item belongs to and calculate its corresponding price and tax
        if item[i] in food:
            price_sum += price[i]* 1.1

        elif item[i] in electronics:
            price_sum += price[i]*1.15

        elif item[i] in liqueur:
            price_sum += price[i]*1.2
            
      # to add the new item in the goods list and caculate its corresponding price and tax
        else:
            cate = input(f"{item[i]} is not in the database. What category of it is in, food, electronics,liqueur?")

            if cate == "food":
                food.append(item[i])
                price_sum += price[i]* 1.1
                
            elif cate == "electronics":
              price_sum += price[i]*1.15
              electronics.append(item[i])
                
            elif cate == "liqueur":
                liqueur.append(item[i])
                price_sum += price[i]*1.2
            
    print(f"The total price for your order is {price_sum} JPY.")
    

total(["bread","beer","ipad"],[300,800,30000])
total(["apple","wine","Mac"],[300,800,150000])
total(["banana","mojito","iphone-12"],[100,1000,80000])
total(["banana","mojito","iphone-11"],[100,1000,70000])
```
