##**Criteria A:Context**
My client, Mr.Smith, is the owner of an apple farm. In conversation with Mr.Smith, he wanted to have a computer system that allows him to record the information of the customers' order and the fees of delivering apples.

At the moment, he is using a monthly spreadsheet to record, which results in several limitations: 
1) repetitive work every month of typing all kinds of information by hand in the spreadsheet;
2) the spreadsheet is so long to read that it's hard for people to trace the data back;
3) Information regarding the product and customer privacy stored in the spreadsheet is not well protected.
Table 1 was provided by Mr.Smith as reference for the deliver fee.

| Distance/km  | Weight/kg | Equations for Deliver fee/$                                   |
|--------------|-----------|-------------------------------------------------|
| 0-3          | ≤10       | 0                                               |
| 3 < d ≤ 1000 | ≤30       | Weight*0.15 + distance * 0.08 - customer coupon |
| 3 < d ≤ 1000 | >30       | Weight*0.3 + distance * 0.08 - customer coupon  |
| d > 1000     | 500 min.     | 230 + (distance-500)*0.08 - customer coupon     |

**Table 1. Equations to calculate deliver fee**

*If the deliver distance is more than 1000km, the farm can still deliver the apples when the weight is more than 500kg.*

###Create class "Apple"

```

class Apple():
  no_apple: int = 0             ## class attribute
  def __init__(apple,color = None,brand = None,weight = None,quality = None,price = None,distance = None,coupon = 0):
    apple.color : str = color
    apple.brand : str = brand
    apple.weight : int = weight
    apple.quality : int = quality
    apple.price : int = price
    apple.distance: int = distance
    apple.coupon: int = coupon 
  #when an object is intantiated an apple is counted, we print a message
    Apple.no_apple += 1
    print("Congratulations! An object apple was created")
  
  def get_info(apple) -> str:
    return f"<Apple information: color: {apple.color},brand: {apple.brand}, weight:{apple.weight}, quality: {apple.quality},price:{apple.price},distance:{apple.distance}>"
  def get_price(apple) -> int:
    return apple.price

  def get_weight(apple)-> int:
    return apple.weight

  def get_quality(apple)-> int:
    return apple.quality

  def get_brand(apple) -> str:
    return apple.brand

  def get_color(apple) -> str:
    return apple.color

  def get_distance(apple) -> int:
    return apple.distance 

  def get_coupon(apple) -> int:
    return apple.coupon 
    
  def set_price(apple, price:int):
    apple.price = price

  def set_quality(apple,quality:int):
    apple.quality = quality

  def set_brand(apple,brand:str):
    apple.brand = brand

  def set_weight(apple,weight:int):
    apple.weight = weight
  

  



  
```

```

def calculate_deliver_fee(weight:int,distance:int,coupon:int):
  if distance <= 3 and weight <= 10:
    return 0
  elif distance == 0:
    return 0
  elif 3<distance<=1000 and weight <=30:
    return(weight*0.15+distance*0.08-coupon) 
  elif 3<distance<=1000 and weight >30:
    return(weight*0.3+distance*0.08-coupon)
  elif distance > 1000 and weight >= 500:
    return(230+ (distance-500)*0.08-coupon) 

for apple in list_apples:
  weight = apple.weight
  distance = apple.distance
  coupon = 0
  print(f" The customer orders the deliver service •••••••••••••••••••• weight: {weight}kg, distance: {distance}km, deliver fee is {calculate_deliver_fee(weight,distance,coupon)}$.")
    

```

###Create class "Customers"

```
class Customers():
  def __init__(self,name,phone_no,email = None,apples_order = None): 
    self.name: str = name
    self.phone_no: int = phone_no
    self.email: str = email
    self.apples_order = []


  def get_name(self):
    return self.name
     
  def add_order(self,weight = 0,myprice=None,distance = 0):
    new_order = Apple(weight=weight,price = myprice,distance=distance)
    self.apples_order.append(new_order)

  def cancel_order(self):
    delete_order = Apple(weight = None,quality = None,price = None,distance = None)
    self.apples_order = self.apples_order[0:-2]
    print("The last order has been cancelled.") 


  def calculate_deliver_fee(self):
    for apple in self.apples_order:
      weight = apple.get_weight()
      distance = apple.get_distance()
      coupon = apple.get_coupon()

      if distance <= 3 and weight <= 10:
        print (f"{self.name} is free for delivery.")
      elif distance == 0:
        print (f"{self.name}: no delivery service.")
      elif 3<distance<=1000 and weight <=30:
        print(f"{self.name}---- orders: {weight}kg, distance: {distance}km, delivery fee: {weight*0.15+distance*0.08-coupon}$.") 
      elif 3<distance<=1000 and weight >30:
        print(f"{self.name}---- orders: {weight}kg, distance: {distance}km, delivery fee: {weight*0.3+distance*0.08-coupon}$.")
      elif distance > 1000 and weight >= 500:
        print(f"{self.name}---- orders: {weight}kg, distance: {distance}km, delivery fee: {230+ (distance-500)*0.08-coupon}$.") 
      else:
        print(f"{self.name}: no delivery service.")
          

  def set_name(self, name:str):
    self.name = name
  def set_phone_no(self, phone_no:int):
    self.phone_no = phone_no
  def set_email(self, email:str):
    self.email = email



```
