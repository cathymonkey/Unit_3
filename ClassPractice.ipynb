```
class Person():
  def __init__(self,name,age):
    self.name: str = name 
    self.age: int = age
  
  def display_name(self) -> str:
    return self.name
  
  def display_age(self) -> int:
    return self.age
  
class Student(Person):
  def __init__(self,name,age,grade):
    Person.__init__(self,name,age)
    self.grade: int = grade

  def display_grade(self) -> int:
    return self.grade
  
  def displayStudent(self):
    print(f"{self.display_name()} is {self.display_age()} years old and is Grade {self.display_grade()}.")
  

s1 = Student("Isabel",17,11)
s1.displayStudent()




```

**2. Write a Python class to convert an integer to a roman numeral.**

```

## create a string first and then by dividing the number to get the roman numeral 
class Roman_num():
    
   def convertToRoman(self,num:int):
    Roman1 = ["I","X","C","M"] # 1,10,100,1000
    Roman2 = ["V","L","D",""] # 5,50,500, ""
    s = ""
    if 0 < num < 4000:
      for i in range(4):
        res = ""
        n = num%10
        if n < 4:
          res += Roman1[i]*n
        elif n == 4:
          res += Roman1[i]+Roman2[i]
        elif 5 <= n < 9:
          res += Roman2[i]+Roman1[i]*(n-5)
        elif n == 9:
          if i < 3:
            res += Roman1[i]+Roman1[i+1]
        s = res + s
        num //= 10
        if num == 0:
          break
    return s

n1 = Roman_num()
print(Roman_num.convertToRoman(n1,19))
print(Roman_num.convertToRoman(n1,24))
print(Roman_num.convertToRoman(n1,36))
print(Roman_num.convertToRoman(n1,44))
print(Roman_num.convertToRoman(n1,57))
print(Roman_num.convertToRoman(n1,100))
print(Roman_num.convertToRoman(n1,549))
print(Roman_num.convertToRoman(n1,2549))


### There might be another way by using modulus without using if.


    


```

**3. Describe with words the program below.**

**Create the UML class diagram for the program below.** 


The program aims to create a bank account and its subaccount -- minimum balance account. Both of the accounts can withdraw and deposit money but there is a limitation on the latter one which doesn't allow to withdraw money when the balance will be smaller than minimum balance the account originally set. The program also shows a one-to-one relationship between the two accounts.

```
class BankAccount:
    def __init__(self):
        self.balance = 0

    def withdraw(self, amount):
        self.balance -= amount
        return self.balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

class MinimumBalanceAccount(BankAccount):
    def __init__(self, minimum_balance):
        BankAccount.__init__(self)
        self.minimum_balance = minimum_balance

    def withdraw(self, amount):
        if self.balance - amount < self.minimum_balance:
            print('Sorry, minimum balance must be maintained.')
        else:
            BankAccount.withdraw(self, amount)

mine = BankAccount()
print(mine.withdraw(20))
print(mine.deposit(100.6))

m2 = MinimumBalanceAccount(10)
print(m2.deposit(100))
m2.withdraw(50)
print(f"The balance now is {m2.balance}.")
m2.withdraw(60)
```
