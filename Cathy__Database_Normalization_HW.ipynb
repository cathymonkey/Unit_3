### **SQL Commands**




```
CREATE TABLE table_name (
    column1 datatype,
    column2 datatype,
    column3 datatype,
   ....
);## Create a table

SELECT * FROM table_name; ## To select data from the table

SELECT DISTINCT column1, column2, ...
FROM table_name; ## To select distinct data from the table

INSERT INTO table_name ## usually no need to write this : (column1, column2, column3, ...)
VALUES (value1, value2, value3, ...); ## To insert values into the table and must be in the same order as of the table's column. 


```
